```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import keras
from keras.utils import np_utils

%matplotlib inline
```

```python
_, (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
from glob import glob

for filename in glob("./*_iter_49*name_block1_conv1*"):
    print(filename)
```

```python
path = "model1_valid_no_fc_batch_norm_21_Adam_0.001_categorical_crossentropy_accuracy_acc_categorical_accuracy_0.1_32_iter_49.h5"
```

```python
model = keras.models.load_model("../restore/"+path)
```

```python
pred = model.predict(x_test)
(np.argmax(pred,axis=1) == np.argmax(np_utils.to_categorical(y_test, 10),1)).mean()
```

```python
i = np.random.randint(x_test.shape[0])
print(i)
```

```python
X_test = np.load("./{}_name_block1_conv1.npy".format(path))
```

```python
print(np.argmax(pred,axis=1)[i], y_test[i])
plt.imshow(x_test[i])
plt.show()
plt.imshow(np.clip(X_test[i],0,1))
plt.title("block1_conv1")
plt.colorbar()
plt.show()
plt.imshow(np.clip(X_test[i][:,:,0],0,1), cmap='gray')
plt.colorbar()
plt.show()
plt.imshow(np.clip(X_test[i][:,:,1],0,1), cmap='gray')
plt.colorbar()
plt.show()
plt.imshow(np.clip(X_test[i][:,:,2],0,1), cmap='gray')
plt.colorbar()
plt.show()
```
